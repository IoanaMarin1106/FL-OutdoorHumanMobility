```python
df = pd.read_csv('places-mobile.csv')
```

```python
df.head()
```

```python
df['surface'] = (df['n_latitude'] - df['s_latitude'])*(df['e_longitude'] - df['w_longitude'])
```

```python
df.sort_values(by='surface', ascending=False).head()
```

```python
center = (44.43934, 26.10434)
map = folium.Map(location = list(center), zoom_start=12)

df_test = df

for i in df_test.index:
    folium.Rectangle(
        bounds=[[df_test['s_latitude'][i], df_test['w_longitude'][i]], [df_test['n_latitude'][i], df_test['e_longitude'][i]]],
        color = '#3186cc',
        fill = True,
        fill_color = '#3186cc'
    ).add_to(map)

map
```
