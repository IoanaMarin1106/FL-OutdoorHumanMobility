```python
import pandas as pd
import numpy as np

np.set_printoptions(threshold=np.inf)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_seq_items', None)
```

```python
df = pd.read_csv('places.csv', dtype={
    'id': 'string',
    'name': 'string',
    'latitude': 'string',
    'longitude': 'string',
    'tags': 'object'
})
```

```python
df.drop_duplicates(subset=['id'], inplace=True)
```

```python
df.shape[0]
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
import ast

df['tags'] = df['tags'].apply(ast.literal_eval)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer(sparse_output=True)

df = df.join(pd.DataFrame.sparse.from_spmatrix(
                mlb.fit_transform(df.pop('tags')),
                index=df.index,
                columns=mlb.classes_)
            )
```

```python
df.head()
```

```python
df.columns
```

```python
df[df['subway_station'] == 1].shape[0]
```

```python
df[(df['cafe'] == 0) & (df['coffee_shop'] == 1)]
```

```python
filtered_df = df[(df['coffee_shop'] == 1) | (df['restaurant'] == 1) | (df['bar'] == 1) | (df['night_club'] == 1)]
```

```python
filtered_df[['id', 'name', 'latitude', 'longitude']].to_csv("filtered-places.csv", index=False)
```

# Work with boundingbox

```python
bdf = pd.read_csv('geocoded-places.csv', dtype={
    'id': 'string',
    'name': 'string',
    'latitude': 'string',
    'longitude': 'string',
    'boundingbox': 'object'
})
```

```python
bdf['boundingbox'] = bdf['boundingbox'].apply(ast.literal_eval)
```

```python
bdf.head()
```

```python
bdf[['s_latitude', 'n_latitude', 'w_longitude', 'e_longitude']] = pd.DataFrame(bdf['boundingbox'].tolist(), index= bdf.index)
```

```python
bdf.drop(axis=1, columns=['boundingbox'], inplace=True)
```

```python
bdf.head()
```

```python
new_df = bdf.drop(axis=1, columns=['id', 'latitude', 'longitude'])
```

```python
new_df[['place_id', 'name', 's_latitude', 'n_latitude', 'w_longitude', 'e_longitude']].to_csv("places-mobile.csv", index=False)
```

```python
new_df[['place_id', 'name', 's_latitude', 'n_latitude', 'w_longitude', 'e_longitude']].to_json("places-mobile.json", orient='records', lines=True)
```
